**FastAPI** - это веб-фреймворк для Python, который позволяет создавать веб-приложения с высокой производительностью и быстрой разработкой. Он использует синхронный код, но благодаря использованию современных технологий, таких как ASGI, FastAPI обеспечивает высокую скорость работы.

Основные преимущества FastAPI:

* Очень быстрый и легковесный;
* Встроенная поддержка Swagger (OpenAPI);
* Отличная поддержка для асинхронных операций;
* Автоматическая валидация данных;
* Простой и интуитивно понятный синтаксис.

**FastAPI** также предоставляет возможность создания асинхронных API, что позволяет обрабатывать множество запросов одновременно и значительно повышает производительность. Кроме того, FastAPI поддерживает автоматическую документацию API с помощью Swagger, что упрощает работу с API для разработчиков.

**FastAPI** является достаточно новым фреймворком, поэтому на данный момент не так много крупных проектов на нем, как на других более устоявшихся фреймворках. Однако, уже есть несколько проектов, которые используют FastAPI:

**Uvicorn** - это ASGI-сервер, который используется для запуска веб-приложений, созданных на основе ASGI-фреймворков, таких как FastAPI, Starlette, Quart и др.

**ASGI** - это асинхронный интерфейс для веб-серверов в Python, который был создан для поддержки асинхронных фреймворков. Он позволяет работать с асинхронным кодом и обрабатывать большое количество запросов одновременно.

Uvicorn обеспечивает высокую скорость и производительность при работе с ASGI-приложениями, благодаря тому, что он использует простой и эффективный асинхронный код на основе asyncio и использованию Event Loop. Также Uvicorn поддерживает многопроцессорный режим работы, что позволяет распределять запросы между несколькими ядрами процессора.

Для запуска приложений, созданных на основе ASGI-фреймворков, необходимо использовать ASGI-сервер, такой как Uvicorn.

````python
from fastapi import FastAPI

app = FastAPI()

@app.get("/")
def read_root():
    return {"Hello": "World"}

@app.get("/items/{item_id}")
def read_item(item_id: int, q: str = None):
    return {"item_id": item_id, "q": q}
````

Это приложение создает экземпляр класса **FastAPI**, который представляет собой веб-приложение. Затем определяются два маршрута - один для корневого пути и один для пути с переменной item_id.

В функции read_root() мы просто возвращаем строку приветствия "Hello, World!". В функции read_item() мы используем параметр item_id для получения значения переменной из URL-адреса и параметр q для получения дополнительного значения из запроса.

````python
script = '''
from fastapi import FastAPI

app = FastAPI()

@app.get("/")
def read_root():
    return {"Hello": "World"}

@app.get("/items/{item_id}")
def read_item(item_id: int, q: str = None):
    return {"item_id": item_id, "q": q}
'''

with open('main.py', 'w') as f:
  f.write(script)
````

* `nohup`: Это утилита в Unix-подобных операционных системах, которая позволяет запустить процесс в фоновом режиме и отключить его от текущего терминала, так что процесс продолжает выполняться, даже если вы выйдете из сеанса или закроете терминал. `nohup` также перенаправляет стандартный вывод и стандартный вывод ошибок в файл `nohup.out`.
* `uvicorn`: Это ASGI (Asynchronous Server Gateway Interface) сервер для Python, который позволяет запускать веб-приложения, написанные с использованием ASGI-совместимых фреймворков.

* `&` помещает процесс в фоновый режим, позволяя вам продолжать работать в командной строке без блокировки ее использования

````python
from fastapi import FastAPI, Request, File, UploadFile  # Импорт необходимых классов и функций из FastAPI
from fastapi.responses import HTMLResponse, StreamingResponse  # Импорт необходимых классов откликов (responses) из FastAPI
from fastapi.staticfiles import StaticFiles  # Импорт класса для обслуживания статических файлов из FastAPI
from fastapi.templating import Jinja2Templates  # Импорт класса шаблонов Jinja2 из FastAPI
from fastapi.middleware.cors import CORSMiddleware  # Импорт класса промежуточного программного обеспечения (middleware) CORS из FastAPI
import uvicorn  # Импорт сервера uvicorn
import io  # Импорт модуля для работы с потоками ввода-вывода (I/O)
import os  # Импорт модуля для работы с функциями операционной системы
import base64  # Импорт модуля для работы с кодированием/декодированием base64
from PIL import Image, ImageOps  # Импорт классов изображений и операций с изображениями из библиотеки PIL (Pillow)
from tensorflow.keras.models import load_model  # Импорт функции для загрузки модели из библиотеки Keras
from tensorflow.keras.preprocessing.image import img_to_array  # Импорт функции для преобразования изображения в массив из библиотеки Keras
import numpy as np  # Импорт библиотеки NumPy для работы с массивами
from typing import Dict  # Импорт класса Dict для указания типа словаря
from pydantic import BaseModel  # Импорт базовой модели (BaseModel) из Pydantic


app = FastAPI()
app.mount("/static", StaticFiles(directory="static"), name="static")

#для google colab
#app.mount("/static", StaticFiles(directory="/content/static"), name="static")

templates = Jinja2Templates(directory="templates")

# Разрешаем запросы CORS от любого источника
origins = ["*"]  # Для простоты можно разрешить доступ со всех источников
app.add_middleware(
    CORSMiddleware,
    allow_origins=origins,
    allow_credentials=True,
    allow_methods=["*"],
    allow_headers=["*"],
)

model_mnist = load_model('static/model_mnist.h5')
model_segment = load_model('static/segment.h5')

@app.get("/")
@app.get("/index.html", response_class=HTMLResponse)
async def home(request: Request):
    return templates.TemplateResponse("index.html", {"request": request})

@app.get("/mnist.html", response_class=HTMLResponse)
async def mnist(request: Request):
    return templates.TemplateResponse("mnist.html", {"request": request})

@app.get("/segment.html", response_class=HTMLResponse)
async def mnist(request: Request):
    return templates.TemplateResponse("segment.html", {"request": request})

@app.get("/tseries.html", response_class=HTMLResponse)
async def mnist(request: Request):
    return templates.TemplateResponse("tseries.html", {"request": request})




@app.post("/mnist_predict")
async def predict(data: Dict[str, str]):
    # Декодируем данные изображения из формата base64
    image_data = base64.b64decode(data['image_data'])
    # Преобразуем данные в объект изображения
    image = Image.open(io.BytesIO(image_data))
    # Преобразуем изображение в массив NumPy и изменяем его размер
    image = image.convert('L').resize((28, 28))
    # Нормализуем значения пикселей изображения
    image = np.array(image) / 255.0
    # Разворачиваем массив в одномерный вектор
    processed_image = np.expand_dims(image.flatten(), axis=0)
    # Используем модель для предсказания цифры
    predictions = model_mnist.predict(processed_image)
    # Получаем номер цифры с наибольшей вероятностью
    digit = np.argmax(predictions)
    # Возвращаем предсказанную цифру
    return {"digit": int(digit)}

@app.post('/segment_predict')
async def segment_predict(image: UploadFile = File(...)):
    img_or = Image.open(image.file).resize((456, 256))
    img = np.array(img_or)
    processed_image = np.expand_dims(img, axis=0)
    predict = np.argmax(model_segment.predict(processed_image), axis=-1)
    mask = labels_to_rgb(predict[..., None])[0]
    # Создание канала альфа-слоя для изображения
    alpha = Image.fromarray(mask).convert('L').point(lambda x: 0 if x == 0 else 255)
    alpha.save('result.png')
    # Применение маски к изображению
    result = img_or.copy()
    result.putalpha(alpha)

    # Создание нового изображения и вставка на него оригинального изображения и изображения с наложенной маской
    new_img = Image.new('RGBA', (img_or.width*2, img_or.height), (0, 0, 0, 0))
    new_img.paste(Image.fromarray(mask), (0, 0))
    new_img.paste(result, (img_or.width, 0))


    buffer_inverted = io.BytesIO()
    new_img.save(buffer_inverted, format="PNG")
    buffer_inverted.seek(0)

    return StreamingResponse(buffer_inverted, media_type="image/png")

def labels_to_rgb(image_list
                 ):
    AIRPLANE = (255, 255, 255)      # Самолет (белый)
    BACKGROUND = (0, 0, 0)          # Фон (черный)
    CLASS_LABELS = (AIRPLANE, BACKGROUND)
    result = []
    for y in image_list:
        temp = np.zeros((256, 456, 3), dtype='uint8')
        for i, cl in enumerate(CLASS_LABELS):
            temp[np.where(np.all(y==i, axis=-1))] = CLASS_LABELS[i]
        result.append(temp)
    return np.array(result)


if __name__ == '__main__':
    uvicorn.run("main:app", host="0.0.0.0", port=int(os.environ.get('PORT', 8000)))
````

Этот код добавляет промежуточное ПО (middleware) CORSMiddleware в приложение FastAPI.

**CORS** расшифровывается как "Cross-Origin Resource Sharing" и является механизмом, который позволяет веб-страницам получать ресурсы с другого домена (origin). Обычно браузеры блокируют такой доступ по соображениям безопасности.

Промежуточное ПО CORSMiddleware обрабатывает запросы и ответы, чтобы обеспечить безопасный обмен данными между сервером и клиентом в сценариях, когда они имеют разные источники.

В данном конкретном коде мы указываем параметры конфигурации для CORSMiddleware:

* allow_origins=["*"] разрешает запросы со всех источников (доменов) (в реальном проекте следует использовать более ограниченные правила для безопасности).

* allow_credentials=True разрешает передачу cookies и заголовков авторизации в запросах.

* allow_methods=["*"] разрешает использование всех HTTP-методов (GET, POST, PUT, DELETE и т.д.).

* allow_headers=["*"] разрешает использование всех HTTP-заголовков.

**Mnist.html**

````python
<script>
		const canvas = document.getElementById('canvas');
		const ctx = canvas.getContext('2d');
		ctx.fillStyle = 'white';
		ctx.strokeStyle = 'white';
		let painting = false;

		function startPosition(e) {
			painting = true;
			draw(e);
		}

		function finishedPosition() {
			painting = false;
			ctx.beginPath();
		}

		function draw(e) {
			if (!painting) return;
			ctx.lineWidth = 14;
			ctx.lineCap = 'round';

			ctx.lineTo(e.clientX - canvas.offsetLeft, e.clientY - canvas.offsetTop);
			ctx.stroke();
			ctx.beginPath();
			ctx.moveTo(e.clientX - canvas.offsetLeft, e.clientY - canvas.offsetTop);
		}

		canvas.addEventListener('mousedown', startPosition);
		canvas.addEventListener('mouseup', finishedPosition);
		canvas.addEventListener('mousemove', draw);

		const clearButton = document.getElementById('clear-button');
		clearButton.addEventListener('click', () => {
			ctx.clearRect(0, 0, canvas.width, canvas.height);
		});

		document.getElementById('predict-button').addEventListener('click', () => {
            const image = canvas.toDataURL();
            const data = image.split(',')[1];
            fetch('https://73dd-35-237-86-171.ngrok-free.app/mnist_predict', {
                method: 'POST',
                headers: {
                    'Content-Type': 'application/json'
                },
                body: JSON.stringify({ image_data: data })
            })
            .then(response => response.json())
            .then(data => {
				const animatedParagraph = document.getElementById("prediction-text");
				animatedParagraph.style.opacity = "1";
                animatedParagraph.textContent = `Распознана цифра: ${data.digit}`;
				animatedParagraph.style.display = "block";
				setTimeout(function() {
					// Плавно скрываем элемент
					animatedParagraph.style.opacity = "0";
				}, 1000);
            });
        });
	</script>
````

Данный код является скриптом JavaScript, который выполняет несколько функций связанных с рисованием на холсте (canvas) и распознаванием написанной цифры с помощью модели машинного обучения. Разберем его построчно:

1. `const canvas = document.getElementById('canvas');` - Здесь создается переменная `canvas`, которая получает элемент холста из HTML-разметки по его идентификатору "canvas".

2. `const ctx = canvas.getContext('2d');` - Создается контекст рисования 2D для холста. Контекст `ctx` используется для выполнения рисования и манипуляций с элементами на холсте.

3. `ctx.fillStyle = 'white';` и `ctx.strokeStyle = 'white';` - Устанавливают цвет заливки и цвет контура для рисования на холсте в белый цвет.

4. `let painting = false;` - Создается переменная `painting`, которая указывает на состояние рисования. Изначально она установлена в значение `false`, что означает, что рисование не выполняется.

5. `function startPosition(e) { ... }` - Объявляется функция `startPosition`, которая устанавливает переменную `painting` в `true` и начинает рисование, вызывая функцию `draw(e)`.

6. `function finishedPosition() { ... }` - Объявляется функция `finishedPosition`, которая устанавливает переменную `painting` в `false` и сбрасывает текущий путь рисования.

7. `function draw(e) { ... }` - Объявляется функция `draw`, которая выполняет рисование на холсте в зависимости от положения курсора. Если переменная `painting` равна `false`, функция завершает работу. В противном случае, функция устанавливает параметры рисования (толщина линии, тип конца линии) и рисует линию от предыдущей точки до текущих координат курсора на холсте. Затем функция начинает новый путь рисования с текущих координат.
